```python
#part1
#multistep method
import numpy as np
import matplotlib.pyplot as plt
def f(y):
    return -y


def solve_euler(y0,h):
    
    y1 = y0+ (h*f(y0))
    return y1

def f1(y0,y1,h):
    
    y2 = (-4*y1) + (5*y0) + h*((4*f(y1)+(2*f(y0))))
    return y2

def solve_Y(h):
    #T = 10 
    #h = 0.1
    n = int(10/h)
    T = np.linspace(0,10,n)
    y_0 = 1
    y_1 = solve_euler(y_0,h)
    Y = [y_0,y_1]
    for i in range (2,n):
        y0 = Y[i-2]
        y1 = Y[i-1]
        y2 = f1(y0,y1,h)
        Y.append(y2)
        
        
    return T,Y

# Plot data on each subplot
T1,Y1 = solve_Y(1)
T2,Y2 = solve_Y(0.1)
T3,Y3 = solve_Y(0.01)
T4,Y4 = solve_Y(0.001)
fig, axs = plt.subplots(2, 2, figsize=(7,6))
axs[0, 0].plot(T1,Y1,color='purple')
axs[0, 0].set_title('h = 1')


axs[0, 1].plot(T2,Y2,color='purple')
axs[0, 1].set_title('h=0.1')


axs[1, 0].plot(T3,Y3,color='purple')
axs[1, 0].set_title('h=0.01')

axs[1, 1].plot(T4,Y4,color='purple')
axs[1, 1].set_title('h=0.001')

for ax in axs.flat:
    ax.set(xlabel='t', ylabel='y')
    
    
fig.suptitle("\nmulti-step method:\n\n computed solution vs time\n\n")
plt.tight_layout()
plt.figure()

fig, axs = plt.subplots(2, 2, figsize=(7,6))
axs[0, 0].plot(T1,(np.exp(-T1)-Y1),color='maroon',linestyle='dotted')
axs[0, 0].set_title('h = 1')


axs[0, 1].plot(T2,(np.exp(-T2)-Y2),color='maroon',linestyle='dotted')
axs[0, 1].set_title('h=0.1')


axs[1, 0].plot(T3,(np.exp(-T3)-Y3),color='maroon',linestyle='dotted')
axs[1, 0].set_title('h=0.01')

axs[1, 1].plot(T4,(np.exp(-T4)-Y4),color='maroon',linestyle='dotted')
axs[1, 1].set_title('h=0.001')

for ax in axs.flat:
    ax.set(xlabel='t', ylabel='Δ')
    
fig.suptitle("\nmulti-step method:\n\n difference (Δ) b/w true and computed solution vs time\n\n")

plt.legend()
plt.tight_layout()
plt.show()    
```

```python
#part 2
#RK method plot

import numpy as np
import matplotlib.pyplot as plt
def f(y):
    return -y

def f_RK(y,h):
    
    k1 = y
    k2 = y + (h/2)*f(k1)
    k3 = y + (h/2)*f(k2)
    k4 = y + h*f(k3)
    yn = y + (h/6)*(f(k1)+(2*f(k2))+(2*f(k3))+f(k4))
    return yn

def solve_RK(h):
    #T = 10 
    #h = 0.1
    n = int(10/h)
    T = np.linspace(0,10,n)
    y_0 = 1
    Y = [y_0]
    for i in range (1,n):
        
        t1 = T[i]
        y0 = Y[i-1]
        y1 = f_RK(y0,h)
        Y.append(y1)
        
        
    return T,Y

# Plot data on each subplot
T1,Y1 = solve_RK(1)
T2,Y2 = solve_RK(0.1)
T3,Y3 = solve_RK(0.01)
T4,Y4 = solve_RK(0.001)
fig, axs = plt.subplots(2, 2, figsize=(7,6))
axs[0, 0].plot(T1,Y1,color='purple')
axs[0, 0].set_title('h = 1')


axs[0, 1].plot(T2,Y2,color='purple')
axs[0, 1].set_title('h=0.1')


axs[1, 0].plot(T3,Y3,color='purple')
axs[1, 0].set_title('h=0.01')

axs[1, 1].plot(T4,Y4,color='purple')
axs[1, 1].set_title('h=0.001')

for ax in axs.flat:
    ax.set(xlabel='t', ylabel='y')
    
    
fig.suptitle("\nRK method:\n\n computed solution vs time\n\n")
plt.tight_layout()
plt.figure()

fig, axs = plt.subplots(2, 2, figsize=(7,6))
axs[0, 0].plot(T1,(np.exp(-T1)-Y1),color='maroon',linestyle='dotted')
axs[0, 0].set_title('h = 1')


axs[0, 1].plot(T2,(np.exp(-T2)-Y2),color='maroon',linestyle='dotted')
axs[0, 1].set_title('h=0.1')


axs[1, 0].plot(T3,(np.exp(-T3)-Y3),color='maroon',linestyle='dotted')
axs[1, 0].set_title('h=0.01')

axs[1, 1].plot(T4,(np.exp(-T4)-Y4),color='maroon',linestyle='dotted')
axs[1, 1].set_title('h=0.001')

for ax in axs.flat:
    ax.set(xlabel='t', ylabel='Δ')
    
fig.suptitle("\nRK method:\n\n difference (Δ) b/w true and computed solution vs time\n\n")

plt.legend()
plt.tight_layout()
plt.show()    
    
    
    
    

```

```python
#part 3
#adams bashforth method
import numpy as np
import matplotlib.pyplot as plt
def f(y):
    return -y


def solve_euler(y0,h):
    
    y1 = y0+ (h*f(y0))
    return y1

def f1(y0,y1,h):
    
    y2 = y1 + (h*(3/2)*f(y1))-(h*(1/2)*f(y0))
    return y2

def solve_AB(h):
    #T = 10 
    #h = 0.1
    n = int(10/h)
    T = np.linspace(0,10,n)
    y_0 = 1
    y_1 = solve_euler(y_0,h)
    Y = [y_0,y_1]
    for i in range (2,n):
        y0 = Y[i-2]
        y1 = Y[i-1]
        y2 = f1(y0,y1,h)
        Y.append(y2)
        
        
    return T,Y



# Plot data on each subplot
T1,Y1 = solve_AB(1)
T2,Y2 = solve_AB(0.1)
T3,Y3 = solve_AB(0.01)
T4,Y4 = solve_AB(0.001)

fig, axs = plt.subplots(2, 2, figsize=(7,6))
axs[0, 0].plot(T1,Y1,color='purple')
axs[0, 0].set_title('h = 1')


axs[0, 1].plot(T2,Y2,color='purple')
axs[0, 1].set_title('h=0.1')


axs[1, 0].plot(T3,Y3,color='purple')
axs[1, 0].set_title('h=0.01')

axs[1, 1].plot(T4,Y4,color='purple')
axs[1, 1].set_title('h=0.001')

for ax in axs.flat:
    ax.set(xlabel='t', ylabel='y')
    
    
fig.suptitle("\nAdams Bashford method:\n\n computed solution vs time\n\n")
plt.tight_layout()
plt.figure()

fig, axs = plt.subplots(2, 2, figsize=(7,6))
axs[0, 0].plot(T1,(np.exp(-T1)-Y1),color='maroon',linestyle='dotted')
axs[0, 0].set_title('h = 1')


axs[0, 1].plot(T2,(np.exp(-T2)-Y2),color='maroon',linestyle='dotted')
axs[0, 1].set_title('h=0.1')


axs[1, 0].plot(T3,(np.exp(-T3)-Y3),color='maroon',linestyle='dotted')
axs[1, 0].set_title('h=0.01')

axs[1, 1].plot(T4,(np.exp(-T4)-Y4),color='maroon',linestyle='dotted')
axs[1, 1].set_title('h=0.001')

for ax in axs.flat:
    ax.set(xlabel='t', ylabel='Δ')
    
fig.suptitle("\nAdams Bashford method:\n\n difference (Δ) b/w true and computed solution vs time\n\n")

plt.legend()
plt.tight_layout()
plt.show()    
    
    
```
